```python
!pip install gTTS -q
from gtts import gTTS
import IPython.display as ipd
import time

# --- 外部API連携 (シミュレーション) ---
def check_availability_api(department):
    """
    // TODO: ここに実際の病院予約システムのAPIを連携させる
    // 例: requests.get(f"https://hospital.com{department}")
    """
    # 今はダミーデータを返す
    return ["9:00", "10:30", "14:00"]

# --- LLMエージェントの思考ロジック ---
def hospital_agent_llm(user_input):
    user_input = user_input.lower()

    # 1. インテント（意図）の解析
    if "予約" in user_input:
        # 2. 外部データのチェック (API呼び出しのシミュレーション)
        # // TODO: ユーザーの発話から「診療科」を抽出するNLP処理を追加
        available_slots = check_availability_api("general")

        slots_text = "、".join(available_slots)
        return f"空き状況を確認しました。現在、{slots_text}が予約可能です。ご希望の時間はありますか？"

    elif "10:30" in user_input:
        # // TODO: 予約確定APIを実行する (POSTリクエスト)
        # // check_availability_api で再度空きを確認してから確定させる
        return "10:30で予約を確保しました。他に確認しておきたいことはありますか？"

    elif "ない" in user_input or "大丈夫" in user_input:
        return "承知いたしました。それでは当日お待ちしております。お大事に！"

    else:
        return "ご用件をどうぞ。予約の空き状況を確認することもできます。"

# --- メインループ ---
print("--- 高機能・予約エージェント 起動 ---")

while True:
    user_in = input("あなた: ")
    if user_in.lower() in ["exit", "終了"]: break

    # LLMプロセス
    response = hospital_agent_llm(user_in)

    print(f"ボット: {response}")

    # 音声出力
    tts = gTTS(response, lang='ja')
    tts.save('res.mp3')
    ipd.display(ipd.Audio('res.mp3', autoplay=True))

    # ユーザーが「お大事に」系で終わったか判定してループを抜けるロジック
    if "お大事に" in response:
        print("--- 会話終了 ---")
        break

    time.sleep(2)
```
